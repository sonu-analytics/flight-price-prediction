```python
df['Date_of_Journey'] # ka matlab hai ki DataFrame (df) ke andar ek specific column (Date_of_Journey) ko access karna
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0] #ka matlab hai ki "Date_of_Journey" column me jo date stored hai (in string format), usko split karke sirf day extract karna.
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int) #is command ka matlab hai ki "Date" column ko integer type me convert karna.
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey', axis = 1, inplace=True)#ka matlab hai ki DataFrame (df) se "Date_of_Journey" column ko permanently hata dena.
```
